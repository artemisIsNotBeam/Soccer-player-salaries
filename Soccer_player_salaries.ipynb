Imports

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# i need to import regression and train test split
```

fixing the data

```python
data = pd.read_csv("drive/MyDrive/raw_wages.csv")

data['Salary'] = data['Salary'].str.replace(',', '')
data['Salary'] = data['Salary'].str.replace('p/a','')
data['Salary'] = data['Salary'].str.replace('€','').astype(float)

data['EU National'] = data['EU National'].replace({' Yes         ': 1, ' No          ': 0}).astype(float)

#data['AT Apps'] = data['AT Apps'].map({'-':0})
data['CR'] = data['CR'].str.replace(',', '')

data['Last Trans. Fee'] = data['Last Trans. Fee'].str.replace('€','')
data['Last Trans. Fee'] = data['Last Trans. Fee'].str.replace('M','')
data['Last Trans. Fee'] = data['Last Trans. Fee'].str.replace('K','')
data['Last Trans. Fee'] = data['Last Trans. Fee'].str.replace('-','0').astype(float)

data['AT Apps'] = data['AT Apps'].str.replace('-','0')
data['AT Apps'] = data['AT Apps'].str.strip().astype(float)

#print(data['AT Apps'].unique())
#remove the At apps that are -
justPrem = data[data['Division'].str.contains('Premier')]

justPrem.head(10)
```

```python
y = justPrem['Salary']
x = justPrem[['EU National', 'Caps','AT Apps','CR','Age','Last Trans. Fee']]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

Training

```python
model = LinearRegression()
fited = model.fit(x_train,y_train)
```

```python
print(model.coef_)
```

Loss


```python
y_pred = model.predict(x_test)

print(model.score(x_test,y_test))
print(model.score(x_train,y_train))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("loss")
print(f'Mean absolute error: {mae:.2f}')
print(f'Mean squared error: {mse:.2f}')
print(f'Root mean squared error: {rmse:.2f}')
```
